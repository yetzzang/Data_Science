#1. 모듈

* 클래스 또는 함수를 모아 놓은 파일 
* 모든 파이썬 파일(.py)는 모듈 
```
# 모든 모듈의 전체 함수 가져오기 
import 모듈명
#사용
모듈명.함수명()
```

````python
import mytotal 
````

````python
tot = mytotal.sum(1, 100)
print(tot)
````

```
# 특정 함수만 골라서 가져오기 
from 모듈명 import 함수1, 함수2, ...
#사용
함수명()
```

````python
from mytotal import sum
````

````python
tot = sum(1, 100)
print(tot)
````

```
# 특정 함수만 골라서 가져오기 
from 모듈명 import *
#사용
함수명()
```


````python
from mytotal import *  # 함수를 다 가져옴; 함수를 다 가져옴 BUT 앞에 함수 앞에 모듈명을 안 붙혀도 됨. 그러나 복잡해졌을 시 가독성이 떨어짐 
tot = sum (1, 100)
print(tot)
````

```
# 모듈명 별명 붙이기
import 모듈명 as 별명
# 사용
별명.함수명()  *가장 많이 쓰는 방법
```

````python
import mytotal as mt
tot = mt.sum(1, 100)
print(tot)
````

#2. 패키지

* 모듈의 상위개념
* 모듈이 모여있는 디렉토리(파일들이 모여있는 폴더)

패키지 관리자
* 외부 모듈을 설치 또는 삭제하기 위한 프로그램
* pip 명령을 사용 (파이썬 3.4 이후 파이썬을 설치하면 pip가 함께 설치됨)
* 윈도우 명령 프롬프트 또는 터미널에서 사용
* pip - version 으로 버전을 확인 가능
