### 1-3. 생성자 _init_
* 클래스 인스턴스가 생성될 때 호출함 
* __ init__(self, 변수1, 변수2, ...):
* self 매개변수는 항상 첫번째에 오며, 자기 자신의 객체를 가리킴
* 이름이 꼭 self일 필요는 없지만 관례적으로 self 사용
* 생성지에서는 해당 클래스가 다루는 데이터를 정의 

````python
class Dog:
  def __init__(self):
    print(self, 'init 호출')
````

````python
Rucy = Dog()  #Rucy는 Dog의 객체 그래서 자기자신의 주소 (self), 그리고  'init 호출'이 프린트 됨
print(Rucy)  #Rucy를 프린트 한것이라 주소만 호출 됐음 

PPomi = Dog()
print(PPomi)
````

````python
class Dog:
  def __init__(self):
    print(self, 'init 호출')
    self.name = '이름없음'  #self. 이 없으면 어떤 변수의 대한 name인지 불명확해짐 
    self.age = 0
````

````python
Rucy = Dog()
print(Rucy)
print(Rucy.name)
print(Rucy.age)

PPomi = Dog()
print(PPomi)
print(PPomi.name)
print(PPomi.age)
````

````python
class Dog:
  def __init__(self, name, age, family='포메'):
    self.name = name  
    self.age = age
    self.family = family  #포메가 디폴트 값임, 그래서 안 넣으면 자동으로 포메로 설정 
````

###1-4. 메소드 정의하기
* 멤버함수라고도 하며, 해당 클래스의 객체에서만 호출 가능한 함수 
* 메소드는 객체에서만 호출되며, 해당 객체의 속성에 대한 연산을 행함 
* 객체이름.메소드명() 형태로 호출

````python
class Counter:
  def __init__(self):  #메소드지만 가장 먼저 실행될 메소드 
    self.num = 0
  def increment(self):#메소드; 객체 안에서 활용되는 함수 
    self.num += 1
  def current_value(self):
    return self.num
  def reset(self):
    self.num = 0
  
````

````python
KBbank = Counter()
print(KBbank.num)

KBbank.increment()
KBbank.increment()
KBbank.increment()
print(KBbank.num)

print(KBbank.current_value())

print('대기 인원 : %d' %KBbank.current_value())
````

````python
HanaBank = Counter()
print(HanaBank.num)


HanaBank.increment()
HanaBank.increment()
HanaBank.increment()
HanaBank.increment()
HanaBank.increment()
print(HanaBank.num)

HanaBank.reset()
print(HanaBank.num)

print('대기인원 : %d' %HanaBank.current_value())
````

###1-5. 메소드 타입
* instance method: 객체 형태로 호출되기 때문에 해당 메소드를 호출한 객체에만 영향을 미침
* class method : class로 호출 (함수 선언 위에 @staticmethod라고 표기)



````python
class Math:
  def add(self, x, y):
    return x + y
  def multiply(self, x, y):
    return x*y
  
````

````python
math = Math()
result1 = math.add(10, 5)
print(result1)
result2 = math.multiply(10,5)
print(result2)
````

````python
# 객체를 호출 X
class Math:
  @staticmethod #다름 
  def add(x,y): #self가 없음  #이 메소드가 바로 메모리에 올라감, 그래서 앞에 이름이 Math로 저장이 됨; BUT 이미 올라가 있어서 메모리 차지를 많이 함 
    return x+y
  def multiply(x,y):
    return x*y
````

````python
result1 = Math.add(10,5)  #객체를 만들지 않아도 클래스 이름으로 메소드 호출/사용 
print(result1)
result2 = Math.multiply(10,5)
print(result2)
````
