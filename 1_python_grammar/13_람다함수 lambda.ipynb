#1. 람다 함수 
* 단일문으로 표현되는 익명함수
* 코디상에서 한번만 사용되는 기능이 있을 때, 굳이 함수로 만들지 않고 일회성으로 만들어 사용 
* 함수 또한 클래스를 통해 생성된 객체 

```python
#기존의 함수 만들기 
def square1(x):
  return x ** 2
```

```python
print(square1(4))
```

```python
#람다 함수 사용법 1
#BUT 잘 쓰지 않음 

lambda x: x ** 2
```

```python
square2 = lambda x: x ** 2
```

```python
print(square2(4))
```

```python
# 람다 함수 사용법2
# BUT 잘 쓰지 않음 

(lambda x: x ** 2)(4)
```

```python
def myData(str):
  return len(str.strip())  # .strip() : 앞 뒤 공백 제거 
```

```python
print(myData('apple'))
```

```python
data = ['watermelon     ', '  apple123456   ', '  banana  ', '    melon123']
```

```python
# sorted() : 정렬하는 함수
print(sorted(data))  #['    melon123', '  apple   ', '  banana  ', 'watermelon     '] b/c 공백 떄문에 
```

```python
print(sorted(data, key=myData))  # myData 실행 후 > [10 11 5 8] > sorted 함수 실행 > ['  banana  ', '    melon123', 'watermelon     ', '  apple123456   ']
```

```python
# 어떨때 람다 함수를 쓰나?
# 람다 함수 사용법3

print(sorted(data, key=lambda str : len(str.strip())))
# 한번 사용하기 때문에 메모리 사용 적음
```

###2. filter 함수 

* 람다가 유용하게 사용되는 대표적인 함수
* 특정 조건을 만족하는 요소만 남기고 필터링 시킴 

```python
num = [1, 2, 3, 4, 6, 9, 10, 14, 15, 18, 20]
print(num)

num = filter(lambda n: n%2 == 0, num)  #짝수만 거르기; 한번만 쓸 것이기 때문에 굳이 def 함수 필요 X 
print(num)  #<filter object at 0x7faf450bfa10>; 저장은 되었지만 리스트가 나오는것이 아니라 메모리 저장 주소가 나온다 

num = list(filter(lambda n: n%2 == 0, num))
print(num)
```

#3. map 함수

* 각 원소를 주어진 수식에 따라 변형하여 새로운 값을 반환 

```python
num = [1, 2, 3, 4, 6, 9, 10, 14, 15, 18, 20]

num = list(map(lambda n: n ** 2, num))
print(num)
```

###문제 
* 아래 리스트에 각 원소에 제곱을 시킨 값이 홀수인 경우 리스트로 출력하는 프로그램을 작성해보자 
* num = [1, 2, 3, 4, 6, 9, 10, 14, 15, 18, 20]

```python
num = [1, 2, 3, 4, 6, 9, 10, 14, 15, 18, 20]
result = []

num = list(map(lambda n : n ** 2, num))
for i in num:
  if i%2 != 0:
    result.append(i)
print(result)
```

```python
num = [1, 2, 3, 4, 6, 9, 10, 14, 15, 18, 20]

print(list(filter(lambda n : n%2 == 1, map(lambda n : n **2, num))))
```
