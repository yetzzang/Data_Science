#1. scope

* 변수에 접근할 수 있는 범위
* local : 가장 가까운 함수안 범위
* global : 함수 바깥의 변수 또는 import된 module

```python
num1 = 10      #global

def func1():
  num2 = 20    #local
  print(num2)
```

```python
print(num1)
#print(num2)  #NameError: name 'num2' is not defined b/c 로컬이라서 
```

```python
num1 = 10      #global

def func1():
  num1 = 20    #local
  print(num1)
```

```python
print(num1)  #global
func1()      #local 안에 있는 아이가 찍ㅎㅣㅁ 
```

```python
# locals() : 로컬 변수를 확인해주는 함수 
# globals() : 글로벌 변수를 확인해주는 함수 

num1 = 10      

def func1():
  num1 = 20    
  print('num1 로컬 변수 : ', 'num11' in locals())  #num1이라는 변수가 local인지 아닌지를 확인해주는 함수 
```

```python
func1()
print('num1' in globals())
print('num11' in globals())
```

#2. global 키워드 

* 함수 내부에서 로컬변수가 아닌 글로벌 변수로 사용하게 함 

```python
num1 = 10

def func1():
  print(num1)

def func2(num):
  num1 = num
```

```python
func1()
func2(5)
print(num1)  #5로 바뀌어야 한다고 생각할 수 있지만 func2 안에 있는 num1은 로컬 변수임. 그래서 글로벌 num1은 바뀌지 않음. locally create num1 variable in func2 
```

```python
num1 = 10

def func1():
  print(num1)

def func2(num):
  global num1
  num1 = num
```

```python
func1()
func2(5)
print(num1)  # global  num1을 create 했기 때문에 gloabl 변수가 locally하게 작용 
```
