#1. 변수 variable

* 변수란 처리할 데이터를 저장시키는 기억장소 

```python
name = '김사과' # =는 우측에 있는 값을 왼쪽의 것에 대입 
age = 20
'''
stack은 사라지는 영역; heap은 사라지지 않는 영역
name이라는 방을 stack에 만듬 > 그리고 heap에도 설치함 > 그리고 heap의 방의 이름은 '김사과'가  됨 > 그리고 그 '김사과'방은 heap의 1000번지에 저장이 되어 있음 
> stack의 방안에 저장된 기능은 only 1000번지를 가르킴 > 이때 1000번지를 가르킬 때는 '김사과'만을 가르키는 것이 아니라 1000번지의 '김사과'라는 방안에 김사과의 기능까지 같이 가르키는 것 
'''
print(name)
print(age)
'''
왜  variable일까? 왜냐면 언제든지 값이 바뀔 수 있기 때문 
'''
name = '반하나'
print(name)
'''
이경우엔 heap에 '김사과'라는 방은 그대로 존재 > 대신 '반하나'라는 새로운 방이 2000번지에 개설 > 그렇게 됐을 때 name은 2000번지를 가르킴 
김사과가 사라지는 것이 아니라 반하나라는 방이 새롭게 개설 되는 것 
'''
```

#3. 변수의 데이터 타입

* 데이터 타입(자료형) : 자려가 저장되는 형태를 의미 

```python
name = '김사과'
age = 20
weight = 50.5
isLover = False 

#type() : 변수의 데이터 타입을 반환해주는 함수 
print (type(name)) #<class 'str'> : 문자형 타입
print(type(age)) #<class 'int'> : 정수형 타입
print(type(weight)) #<class 'float'> : 실수형 타입 
print(type(isLover)) #<class 'bool'> : 논리형 타입 
```
