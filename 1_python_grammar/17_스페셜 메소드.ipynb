#1. 스페셜 메소드 (Special Method)

* __로 시작해서 __로 끝나는 특수함수
* 해당 메소드들을 구현하면 객체에 여러가지 파이썬 내장 함수나 연산자에 원하는 기능을 적용 가능 

```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def print_pt(self):
    print('({}, {})'.format(self.x, self.y))

  def __str__(self):  # str() 함수를 오버라이딩 ;기능: 1. int나 다른 기호들을 스트링으로 바꿔줌  2. 객체를 넣게 되면 memory 주소를 찍어주는 역할; 현재는 좌표를 찍도록 오버라이딩 했음  
    return ('({}, {})'.format(self.x, self.y))

  def __add__(self, pt):  # + 파이썬 함수/연산자
    new_x = self.x + pt.x
    new_y = self.y + pt.y
    return Point(new_x, new_y)

  def __sub__(self, pt):  # - 파이썬 함수/ 연산자 
    new_x = self.x - pt.x
    new_y = self.y - pt.y
    return Point(new_x, new_y)

  def __mul__(self, m):  # * 파이썬 연산자 
      return Point(self.x * m, self.y * m)

  def __len__(self):  #string의 길이를 주는 파이썬 연산자 
    return self.x * self.y

  def __getitem__(self, index):  # 인덱스 가져오기; 예) 객체[0]  
    if index == 0:
      return self.x
    elif index == 1:
      return self.y
    else:
      return -1 
```

```python
p1 = Point(3, 4)
print(str(p1))
```

```python
p2 = Point(5,8)
```

```python
p3 = p1 + p2  # 객체 덧셈?? __add__ 를 오버라이딩 함; pt2의 x,y값은 pt parameter 됨 
print(p3)  # print(str(p3))와 같은 형상이 됨
```

```python
p4 = p1 - p3
print(p4)
```

```python
p5 = p1 * 5
print(p5)
```
