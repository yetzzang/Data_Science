* 랜덤으로 1또는 7 을 7번출력한다. 7이 연속으로 3번이상이면 "당첨" 아니면 "꽝" 출력하시오.
> 
[예시    ]
7 7 1 1 1 7 1 :     꽝
1 7 7 7 1 1 7  당첨


(11:19 - 11:31)

````python
import random

random_nums = []
for i in range(7):
    random_num = random.choice([1,7])
    random_nums.append(random_num)

print(list)

sum = []
for i in range(len(random_nums[:-2])):
    x = random_nums[i]
    y = random_nums[i+1]
    z = random_nums[i+2]
    sum.append(x+y+z)

if 21 in sum:
    print('당첨')
else:
    print('꽝')

# 메모리를 줄이기 Hint: 리스트를 쓰지 않고 
    # count라는 변수 생성 -> 7이면 + 아니면 _
# 7이 100번이더라도 대응할 수 있게끔 
````

````python
import random

seven_count = 0
win = False
for i in range(7):
    random_num = random.choice([1,7])
    if random_num == 7: 
        seven_count += 1
    else:
        seven_count = 0
    if seven_count == 3:
        win = True
    print(random_num, end = ' ')
if win == True:
    print('당첨')
else:
    print('꽝')
````


1. **`random()`**: 0.0과 1.0 사이의 부동 소수점 수를 반환합니다.
    ```python
    import random
    rand_num = random.random()
   ```

2. **`randint(a, b)`**: a와 b 사이 (둘 다 포함)의 정수를 반환합니다.
    ```python
    import random
    rand_int = random.randint(1, 10)
    ```

3. **`uniform(a, b)`**: a와 b 사이의 부동 소수점 수를 반환합니다.
    ```python
    import random
    rand_float = random.uniform(1.0, 10.0)
    ```

4. **`choice(seq)`**: 시퀀스(seq)에서 랜덤하게 한 항목을 선택합니다.
    ```python
    import random
    my_list = [1, 2, 3, 4, 5]
    rand_choice = random.choice(my_list)
    ```

5. **`shuffle(seq)`**: 시퀀스(seq)의 항목들을 랜덤하게 섞습니다. (원본이 변경됩니다)
    ```python
    import random
    my_list = [1, 2, 3, 4, 5]
    random.shuffle(my_list)
    ```

6. **`sample(population, k)`**: population에서 중복되지 않는 k개의 항목을 랜덤하게 선택하여 리스트로 반환합니다.
    ```python
    import random
    my_list = [1, 2, 3, 4, 5]
    random_sample = random.sample(my_list, 3)
    ```

* 컴퓨터랑 가위바위보해서 결과 출력
> 
내가 이기면 승리
지면 패배
비기면 무승부


(11:45 - 11:51)

````python
import random 

rsp = ['가위', '바위', '보']
computer = random.choice(rsp)
user = input('가위, 바위, 보 중 하나를 내세요:' )

# 가위 win over 보, 바위 win over 가위, 보 win over 바위
if computer == user:
    result = '무승부'
elif (computer == '가위' and user == '보') or (computer == '바위' and user == '가위') or (computer == '보' and user == '바위'):
    result = '패배'
else:
    result = '승리'
    
print(f'컴퓨터: {computer} 사용자:{user} 결과:{result}')
````

* 두 정수를 입력 받고
두 수중에 더 큰 수를 출력
같으면 두 수 모두 출력

(1:10 - )

````python
def get_first():
    first = input('첫번째 정수를 입력하세요:')
    try:
        first = int(first)
    except Exception:
        print('정수를 입력하세요')
        first = get_first()
    return first 

def get_second():
    second = input('두번째 정수를 입력하세요:')
    try:
        second = int(second)
    except Exception:
        print('정수를 입력하세요')
        second = get_second()
    return second

first = get_first()
second = get_second()

if first > second:
    print(first)
elif first == second:
    print(first, second)
else:
    print(second)
````

* 학점 출력 프로그램 작성

-~100 사이의 점수를 입력 받
고
-1~100은 A-81~90은 B-
71~80은 -
61~70- D
0~60은 F 를 
출력하세요!
그 밖에 점수는 비정상적인 점수라서 다시 입력하라고 하세요!


````python
if score>90:
    print('A')
elif score >80:
    print('B')

# 간단한게 코드 덜쓰면서, if 개념 원리에 접근 
````

````python
grades = {'A':None, 'B':None, 'C':None, 'D':None}
for (i, grade) in zip(range(100, 59,-10), grades):
    grades[grade] = [i for i in range(i, i-10, -1)]
grades['F'] = [i for i in range(0,61)]

def get_score():
    score = input('점수를 입력하세요')
    try:
        score = int(score)
        if score < 0 or score > 100:
            print('0~100사이의 점수가 필요합니다')
            score = get_score()
    except Exception:
        print('점수를 다시 입력하세요')
        score = get_score()
    return score 

score = get_score()

for grade, scores in grades.items():
    if score in scores:
        print(f'점수: {score} 학점: {grade}')
   
````

* 사용자에게 숫자를 입력받고 2의 배수인지 아닌지를 출력하는 프로그램을 만드시오
* 단, 사용자에게 입력받은 것이 숫자가 아닐경우 "숫자가 아닙니다"를 출력하시오.


````python
def get_input():
    user_input = input('숫자를 입력하세요:')
    try:
        user_input = int(user_input)
    except Exception:
        print('숫자가 아닙니다')
        user_input = get_input()
    return user_input

user_input = get_input()

if user_input % 2 == 0:
    print(f'{user_input}은 2의 배수입니다')
else:
    print(f'{user_input}은 2의 배수가아닙니다')
````
