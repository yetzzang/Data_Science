* 1부터 100까지 숫자중에 3의 배수를 찾아서 출력 (1분이내)

```python
n = 1
while n <= 100:
    if n % 3 == 0:
        print(n)
    n += 1
```

* 구구단 1~9단까지 출력해보기 (9:05-9:07)

```python
n = 1
while n <= 9:
    print(f'{n}단')
    for i in range(11):
        print(f'{n} x {i} = {n * i}')
    n += 1
```

* 무한 반복으로 숫자를 입력받아 해당 구구단 출력하기* 
2를 입력받았으면 2단만 출력, -1을 입력하면 무한 반복 종
  
(9:08 - 9:33)


```python
while True:
    user_input = input('구구단 출력을 위한 원하는 숫자를 입력하세요. 종료를 원한다면 -1을 입력하세요:')
    user_input = int(user_input)
    if user_input == -1:
        print('구구단 출력이 종료됩니다')
        break
    else: 
        print(f'{user_input}단')
        i = 0
        while i <= 10:
            print(f'{user_input} x {i} = {user_input * i}')
            i += 1
```

(9:35-9:41) (9:42-9:44)

```python
for n in range(9,1,-1):
    print(f'{n}단')
    for i in range(9,-1,-1):
        print(f'{n} x {i} = {n*i}')
```

> 시간복잡도 O(n^2)

```python
n = 9
while True:
    if n < 2:
        break
    print(f'{n}단')
    for i in range(9, 0, -1):
        print(f'{n} x {i} = {n*i}')
    n -= 1
```

> range(start,end,interval)

> start: starting number

> end: ending number -1

> interval: interval of number, if is negative: reversed 

* 369게임 
글자를 무한대로 입력받아서 현재 차례에 맞지않는 내용을 입력하면 게임 료

(9:49-10:04)


```python
print('369게임을 시작합니다')

# 3,6,9가 들어가면 '짝'
# 숫자는 +1씩 업데이트
# 33은? 

answer_num = 0
answer = '0'
answer_list = []
while True:

    # answer 업데이트
    answer_num += 1
    answer = str(answer_num)
    answer_list = [char for char in answer]
    if '3' in answer_list or '6' in answer_list or '9' in answer_list:
        answer = '짝'
    
    # 값을 입력받기 
    user_input = input('다음에 해당하는 값을 입력하세요:')

    # 값이 틀리다면 
    if user_input != answer:
        print('땡! 게임을 종료합니다.')
        break
```
