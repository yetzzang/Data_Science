#1. 큐의 구조 Queue

* 가장 먼저 넣은 데이터를 가장 먼저 꺼낼 수 있는 구조 
* 줄을 서는 행위와 유사
* FIFQ (First In First Exit) 

### 1-1. 큐의 사용 

* 멀티테스킹을 위한 프로세스 스케쥴링 방식을 구현 (운영 체제)
* 푸시메세지
* 예시) 카카오톡 메세지 주고 받기; 먼저 보낸것이 먼저 가감 

###1-2 큐의 용어
* Enqueue: 큐의 데이터를 넣는 기능
* Dequeue: 큐의 데이터를 꺼내는 기능 
* [Visualgo](https://visualgo.net/en/list)

### 1-3 파이썬 queue 라이브러리를 활용하여 queue 자료구조 사용 
* 큐 예시: 집을 지을때 아무것도 주지 않고, 도구만 줌; 사람, 벽돌, 흙, 콘크리트 아무것도 주지 않음. 
* 프레임워크: 이미 지어진 집을 줌. 하지만 안에는 텅텅 비어있음. 파이썬 사용

* Queue(): 가장 일반적인 큐 자료구조를 생성
* LifpQueue(): 나중에 입력된 데이터가 먼저 출력되는 큐의 자료구조를 생성 (스텍구조와 비슷)
* PriorityQueue(): 데이터마다 우선 순위를 넣어서 운선순위가 높은순으로 데이터를 출력 

```python
# 일반적인 Queue 만들기
import queue

data_queue = queue.Queue()
data_queue.put('Hello')  # Enqueue
print(data_queue)  #메모리 주소가 찍힘 
print(data_queue.qsize())
data_queue.put('Python')
print(data_queue.qsize())
print(data_queue.get()) #Dequeue: 값을 리턴 후 삭제
print(data_queue.qsize())
```

```python
data_queue.put(10)
data_queue.put(True)
data_queue.put(100.5)
print(data_queue.qsize())
```

```python
for i in range(data_queue.qsize()):
  item = data_queue.get()
  print(item)
 # 데이터를 다 리턴 후 삭제했기 떄문에 현재 data_queue는 empty
```

```python
# LifoQueue 만들기
data_queue = queue.LifoQueue()
data_queue.put('Hello')
data_queue.put(10)
data_queue.put(True)
print(data_queue.qsize())
print(data_queue.get())  #마지막 입력값이 가장 먼저 리턴 
```

```python
# PriorityQueue 만들기; 튜플로 데이터를 넣음, 번호를 지정해서 넣음 
data_queue = queue.PriorityQueue()
data_queue.put((10, '김사과'))
data_queue.put((5, '반하나'))
data_queue.put((7, '오렌지'))
data_queue.put((8, '이메론'))
data_queue.put((2,'채예리'))
print(data_queue.qsize())
print(data_queue.get())
print(data_queue.get())
print(data_queue.get())
```

###문제
리스트 변수로 큐를 다루는 enqueue, dequeue 기능을 구현해보자 

```python
queue_list = list()

def enqueue(data):
  queue_list.append(data)

def dequeue():
  data = queue_list[0]
  del queue_list[0]
  return data
```

```python
print(len(queue_list))
```

```python
for index in range(10):
  enqueue(index)
  
```

```python
print(len(queue_list))
queue_list
```

```python
print(dequeue())
queue_list
print(dequeue())
queue_list
```
