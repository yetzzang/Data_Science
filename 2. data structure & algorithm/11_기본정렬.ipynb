# 2. 버블 정렬 Bubble Sort

* 두 인접한 데이터를 비교해서 앞에 있는 데이터가 뒤에 있는 데이터보다 크면(작으면) 자리를 바꾸는 정렬 알고리즘
* 특징: element가 n이라고 할때 첫번째 시도 = n, 두번째 시도 = (n-1), 세번째 시도 = (n-2), ... , n번째 시도 = 1만큼 돌게 됨  
<img src="https://upload.wikimedia.org/wikipedia/commons/c/c8/Bubble-sort-example-300px.gif" width=500>

```python
# 오름차순 bubble sort
# if want to have 내림차순, change the inequality sign to less than (<) in if문
def bubblesort(data):
  # let there is n dataset, then len(data) = n, then len(data) -1 = n - 1
  # then it roops from 0 to (n-2)
  for i in range(len(data) - 1):
    # if i = 0, then it roops from 0 to (n-2)
    # if i = 1, then it roops from 0 to (n-3)
    # this means the feature of bubble sort: 
    # first try = n-1, second try = n-2, thrid try = n-3 element comparison between each other 
    for index in range(len(data) - i - 1):
      # if data is bigger than data right to it (data[index + 1]) 
      if data[index] > data[index + 1]: 
        # switch the data 
        data[index], data[index + 1] = data[index + 1], data[index]      
  return data
```

```python
data_list = [80, 90, 100, 70, 50]
bubblesort(data_list)
```

# 3. 삽입 정렬 Insertion Sort 

* 해당 인덱스 (key값) 앞에 있는 데이터(A)부터 비교해서 key값이 더 작으면 (크면), 데이터 (A) 값을 뒤 인덱스로 복사
* key 값이 더 큰 데이터를 만날때까지 반복, 큰 데이터를 만난 위치 바로 뒤에 key값을 이동
* 두번째 인덱스부터 비교를 시작 

<img src="https://upload.wikimedia.org/wikipedia/commons/9/9c/Insertion-sort-example.gif">

```python
def insertion_sort(data):
  # same for loop algorithm as bubble_sort 
  for index1 in range(len(data) -1):
    # but we goes backward from index1
    for index2 in range(index1 + 1, 0, -1):
      if data[index2] < data[index2 - 1]:
        data[index2], data[index2 - 1] = data[index2 - 1], data[index2]
      else:
        break
  return data
```

```python
data_list = [80, 90, 100, 70, 50]
insertion_sort(data_list)
```

# 4. 선택 정렬 Selection Sort

* 주어진 데이터 중, 최소값을 찾음
* 해당 최소값을 데이터 맨 앞에 위차한 값과 교체함
* 맨 앞의 위치를 뺀 나머지 데이터를 동일한 방법으로 반복 

<img src="https://upload.wikimedia.org/wikipedia/commons/9/94/Selection-Sort-Animation.gif" width=100>

```python
# by me 
def selection_sort(data):
  for index in range(len(data)):
    smallest = min(data[index:len(data)])
    min_index = data.index(smallest)
    data[index], data[min_index] = data[min_index], data[index]
  return data
```

```python
data_list = [80, 90, 100, 70, 50]
selection_sort(data_list)
```

```python
# by lecturer
def selection_sort(data):
  for stand in range(len(data) -1):
    min = stand
    for index in range(stand + 1, len(data)):
      if data[min] > data[index]:
        min = index
    data[min], data[stand] = data[stand], data[min]
  return data
```

```python
data_list = [80, 90, 100, 70, 50]
selection_sort(data_list)
```
