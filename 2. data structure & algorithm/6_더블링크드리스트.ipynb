#1. 더블링크드리스트 Doubly Linked List 

* 양방향으로 연결되어 있어서 노트 탐색이 양쪽으로 모두 가능 

노트

* 기본형 링크드리스트 : data + pointer
* 더블 링크드리스트 : prev pointer + data + next pointer 

```python
class Node:
  def __init__(self, data, prev=None, next=None):
    self.prev = prev
    self.data = data
    self.next = next
```

```python
class NodeMgmt:
  def __init__(self, data):
    self.head = Node(data)
    self.tail = self
  
  def insert(self, data):
    if self.head == None:
      self.head = Node(data)
      self.tail = self
    else:
      node = self.head
      while node.next:
        node = node.next
      new = Node(data)
      node.next = new
      new.prev = node
      self.tail = new
  
  def node_print(self):
    node = self.head
    while node:
      print(node.data)
      node = node.next
  
  def insert_before(self, data, after_data): # 2.5, 3
    if self.head == None:
      self.head = Node(data)
      return True
    else:
      node = self.tail
      while node.data != after_data:
        node = node.prev
        if node == None:
          return False
      new = Node(data)
      before_new = node.prev # node.prev는 2를 가리키고 있음
      before_new.next = new # new(2.5) 주소를 원래 3을 가리키고 있던 before_new.next에 저장
      new.prev = before_new # new(2.5) prev에 2의 주소를 저장
      new.next = node # new(2.5) next에 3의 주소를 저장
      node.prev = new # 3의 prev에 new(2.5) 주소를 저장
      return True

# 앞에서부터 검색하여 전달한 data가 있는 경우 해당 값을 return 하고 존재하지 않으면 False return 
  def search_from_head (self, data):
    if self.head == None:
      return False
    node = self.head 
    while node: 
      if node.data == data:
        return node.data
      else: 
        node = node.next 
    return False 

# 뒤에서부터 검색하여 전달한 data가 있는 경우 해당 값을 반환하고 존재하지 않으면 False 반환 
  def search_from_tail(self, data):
    if self.tail == None:
      return False
    node = self.tail 
    while node:
      if node.data == data:
        return node.data
      else:
        node = node.prev 
    return False


  
```

```python
double_linked_list = NodeMgmt(0)

for data in range(1, 11):
  double_linked_list.insert(data)
  
```

```python
double_linked_list.node_print()
```
