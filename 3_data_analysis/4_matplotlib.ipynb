# 2. 스타일 옵션

````python
plt.plot([1,2,3], [3,6,9])
plt.plot([1,2,3], [2,4,9])

plt.title("Titlte test", fontsize=30)

plt.xlabel("X", fontsize=20)
plt.ylabel("Y", fontsize = 20, rotation=90)

plt.show()
````

````python
plt.plot(np.arange(10), np.arange(10)*2)
plt.plot(np.arange(10), np.arange(10)**2)
plt.plot(np.arange(10), np.log(np.arange(10)))

plt.title("title", fontsize=20)

plt.xlabel("X", fontsize=20)
plt.ylabel("Y", fontsize=20, rotation=0)

plt.xticks(rotation=90)
plt.yticks(rotation=30)

# plt.legend(['10*2', '10*3', 'log'], fontsize=15, loc=(0.5, 0.5), ncol=3)
plt.legend(['10*2', '10*3', 'log'], fontsize=15, loc='upper right', ncol=3)

plt.show()
````

# 3. 스타일 세부 설정

````python
plt.plot(np.arange(10), np.arange(10)*2,
         marker='o',
         markersize = 5,
         linestyle='',
         color='blue',
         alpha = 0.3, #투명도
         )
plt.plot(np.arange(10), np.arange(10)**2,
         marker='v',
         markersize = 7,
         linestyle='--',
         color='red',
         alpha = 0.7,
         )
plt.plot(np.arange(10), np.log(np.arange(10)*2),
         marker='*',
         markersize = 15,
         linestyle='-.',
         color='orange',
         alpha=0.5)

plt.title("set marker", fontsize=20)

plt.xlabel("X axis", fontsize=20)
plt.ylabel("Y axis", fontsize=20)

plt.xticks(rotation=90)
plt.yticks(rotation=30)

plt.grid()
plt.show()
````

````python
# note: numpy.random.rand from a uniform distribution (in the range [0,1))
# note: numpy.random.randn generates samples from the normal distribution

x = np.random.rand(50)
y = np.random.rand(50)

# arange(num): 0부터 num - 1까지의 자연수 순서대로 나열
colors = np.arange(50)
area = x * y * 100

print(colors)
print(area)
````

````python
# s = size, c=color
plt.scatter(x,y,s=area, c=colors)
plt.show()
````

````python
# plt.rcParams['figure.figsize'] = (12, 6)
plt.figure(figsize=(12,6))

plt.subplot(1,3,1)
plt.scatter(x,y, s=area, alpha=0.1)
plt.title("alpha=0.1")

plt.subplot(1,3,2)
plt.scatter(x,y,s=area, alpha=0.5)
plt.title('alpha=0.5')

plt.subplot(1,3,3)
plt.scatter(x,y,s=area, alpha=1)
plt.title('alpha=1')

plt.show()
````

````python
x = ["Math", 'Programming', 'Art', 'English', 'Data Science', 'English', "Korean"]
y = [60, 80, 50, 90, 20, 70, 40]

plt.figure(figsize=(6,3))
plt.bar(x, y, align='edge', alpha=0.6, color='red')
plt.xticks(x)
plt.ylabel("Number of Students")
plt.title('Subjects')

plt.show()
````

````python
x = ["Math", 'Programming', 'Art', 'English', 'Data Science', 'English', "Korean"]
y = [60, 80, 50, 90, 20, 70, 40]

plt.figure(figsize=(6,3))
plt.barh(x, y, align='edge', alpha=0.6, color='red')
plt.yticks(x)
plt.ylabel("Number of Students")
plt.title('Subjects')

plt.show()
````
