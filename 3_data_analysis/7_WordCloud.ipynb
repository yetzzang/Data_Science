```python
alice = open('./alice.txt').read()
print(alice)
```

```python
from PIL import Image # 이미지에서 마스킹 데이터를 출력
from wordcloud import STOPWORDS # 불용어, 워드 클라우드로 작성할 단어에서 제외할 단어를 설정
import numpy as np
```

```python
alice_mask = np.array(Image.open('./alice_mask.png'))
alice_mask
```

```python
# 필요 없는 자주 나오는 단어들을 STOPWORDS로 다 묶어버림 ex) the, a, should, is, ...
stopwords = set(STOPWORDS)
print(type(stopwords))
print(stopwords)
```

```python
stopwords.add('said')
stopwords
```

```python
wc = WordCloud(background_color='white', max_words=500, stopwords=stopwords, mask=alice_mask)
wordcloud = wc.generate(alice)
```

```python
plt.figure(figsize=(15,10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off')
plt.show()
```

```python
# 워드 클라우드 만들기
# 1. kolaw의 constitution.txt를 불러오기
# 2. 명사만 추출하기
# 3. stop_words를 아래와 같이 설정
# stop_words =  ['제', '월', '일', '조', '수', '때', '그', '이', '바', '및', '안', '정', '관', '위', '장']
# 4. 명사의 빈도수를 계산해서 총 500개의 단어만 선택
# from collections import Counter
# 5. 위 조건에 맞는 워드 클라우드 만들기
```

```python
from konlpy.corpus import kolaw
```

```python
kolaw_text = kolaw.open('constitution.txt').read()
```

```python
from konlpy.tag import Okt
```

```python
okt = Okt()
tokens_ko = okt.nouns(kolaw_text)
```

```python
stopwords = ['제', '월', '일', '조', '수', '때', '그', '이', '바', '및', '안', '정', '관', '위', '장']
tokens_ko = [each_word for each_word in tokens_ko if each_word not in stopwords]
tokens_ko
```

```python
from collections import Counter
```

```python
count = Counter(tokens_ko)
```

```python
data = count.most_common(500)
tmp_data = dict(data)
tmp_data
```
