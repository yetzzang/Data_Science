```python
import folium
```

```python
# folium.Map(location=[위도, 경도], zoom_start=배율, ..)
map_folium = folium.Map(location=[37.5000425, 127.0359277], zoom_start=17)
# save(): 작성된 지도차트를 html로 저장
map_folium.save('./map.html')
map_folium
```

```python
# 지도 스타일(Open Street Map(기본값), Stamen Terrain, Stamen Toner)
map_folium = folium.Map(location=[37.511835, 127.038549], zoom_start=17, tiles='Stamen Toner')
map_folium
```

```python
# Marker(): location 속성으로 지정한 위치에 마커를 생성
# Popup: 마커를 클릭했을 때 표시할 문자열을 설정
# add_to(): 작성된 마커를 지도에 추가
popup = folium.Popup('코리아 IT 아카데미', max_width=200)
map_folium = folium.Map(location=[37.5000425, 127.0359277], zoom_start=17)
folium.Marker(location=[37.5000425, 127.0359277], popup=popup).add_to(map_folium)
map_folium
```

```python
map_folium = folium.Map(location=[37.56652479270633, 126.9779129116262], zoom_start=15)

popup = folium.Popup('서울특별시청', max_width=200)
folium.Marker(location=[37.56652479270633, 126.9779129116262], popup=popup).add_to(map_folium)

popup = folium.Popup('덕수궁', max_width=200)
folium.Marker(location=[37.56585320879458, 126.97514572197248], popup=popup).add_to(map_folium)

popup = folium.Popup('서울지방경찰청', max_width=200)
folium.Marker(location=[37.57504045678939, 126.97195576938117], popup=popup).add_to(map_folium)

popup = folium.Popup('광화문', max_width=200)
folium.Marker(location=[37.57617987638928, 126.97694467807803], popup=popup).add_to(map_folium)

popup = folium.Popup('영풍문고-종로본점', max_width=200)
folium.Marker(location=[37.56999947440024, 126.98228391170936], popup=popup).add_to(map_folium)

map_folium
```

```python
# icon url : getbootstrap.com/docs/3.3/components/#glyphicons-glyph
# Icon() : 아이콘 모양을 설정

map_folium = folium.Map(location=[37.56652479270633, 126.9779129116262], zoom_start=16)

popup = folium.Popup('서울특별시청', max_width=200)
folium.Marker(location=[37.56652479270633, 126.9779129116262], popup=popup, icon=folium.Icon(color="red", icon="star")).add_to(map_folium)

popup = folium.Popup('덕수궁', max_width=200)
folium.Marker(location=[37.56585320879458, 126.97514572197248], popup=popup, icon=folium.Icon(color="red", icon="star")).add_to(map_folium)

popup = folium.Popup('서울지방경찰청', max_width=200)
folium.Marker(location=[37.57504045678939, 126.97195576938117], popup=popup, icon=folium.Icon(coler="blue", icon='user')).add_to(map_folium)

popup = folium.Popup('광화문', max_width=200)
folium.Marker(location=[37.57617987638928, 126.97694467807803], popup=popup, icon=folium.Icon(color='purple', icon='hand-dwon')).add_to(map_folium)

popup = folium.Popup('영풍문고-종로본점', max_width=200)
folium.Marker(location=[37.56999947440024, 126.98228391170936], popup=popup, icon=folium.Icon(color='green', icon='okay')).add_to(map_folium)

map_folium
```

```python
# CircleMarker(): 원 모양의 마커를 생성

map_folium = folium.Map(location=[37.56652479270633, 126.9779129116262], zoom_start=16)

popup = folium.Popup('서울특별시청', max_width=200)
folium.CircleMarker(location=[37.56652479270633, 126.9779129116262], popup=popup,
                    radius=100, color='#C2185B', fill_color='#EC407A').add_to(map_folium)

map_folium
```
