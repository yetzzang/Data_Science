# 2. basicenglishspeaking.com

```python
# HTML 가져오기
import requests

# HTML 분석툴
from bs4 import BeautifulSoup
```

```python
site = 'https://basicenglishspeaking.com/daily-english-conversation-topics/'
request = requests.get(site)
print(request)
# print(request.text)
html = request.text
```

```python
# persing
soup = BeautifulSoup(html)
```

```python
# css 안에서 해당하는 class name을 찾기
divs = soup.findAll('div', {'class':'thrv-columns'})
print(divs)
```

```python
subject = []

for div in divs:
  links = div.findAll('a')
  # print(links)
  for link in links:
    # print(link.text)
    subject.append(link.text)
```

```python
subject
```

```python
print('총', len(subject), '개의 주제를 찾았습니다')
for i in range(len(subject)):
  print('{0:2d}, {1:s}'.format(i+1, subject[i]))
```

# 3. 뉴스기사

```python
# https://v.daum.net/v/20221006165213133
# https://v.daum.net/v/20221006185601019

def daum_news_title(news_id):
    url = 'https://v.daum.net/v/{}'.format(news_id)
    request = requests.get(url)
    soup = BeautifulSoup(request.text)
    # findAll 해당값 모두 찾아오는거고
    # find 한개만
    title = soup.find('h3', {'class':'tit_view'})
    if title:
        return title.text.strip()
    return ''
```

```python
daum_news_title('20221006165213133')
```

# 4. 벅스뮤직 차트

```python
# music.bugs.co.kr/chart

request = requests.get('https://music.bugs.co.kr/chart')
soup = BeautifulSoup(request.text)

titles = soup.findAll('p', {'class':'title'})
artists = soup.findAll('p', {'class':'artist'})
```

```python
for i in range(len(titles)):
  title = titles[i].text.strip()
  # 공백을 제거 뒤에 나눠줌
  # \n 은 enter를 뜻함
  # [0]하면 앞에것만 추출
  artist = artists[i].text.strip().split('\n')[0]
  data = '{0:3d}위 {1} - {2}'.format(i+1, artist, title)
  print(data)
```

```python
import time

genie = []

header = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) '}
url = 'https://genie.co.kr/chart/top200?ditc=D&ymd=20230712&hh=07&rtm=Y&pg='

for i in range(1,5):
  site = '{}{}'.format(url, i)
  request = requests.get(site, headers=header)
  # for loop이 워낙 빠르기 때문에 internet access에서 만들어내는 시간을 벌어주기 위해
  time.sleep(1)
  soup = BeautifulSoup(request.text)

  # html에서 띄어쓰기는 다른 제목으로 간주하기 때문에
  # 띄어쓰기가 되지 않은 하나만 들고옴
  # <a> 테이블에 title class가 많이 존재함 > 그래서 필요없는 부분도 추출됨
  # titles = soup.findAll('a', {'class':'title'})
  # artists = soup.findAll('a', {'class':'artist'})

    # 새로운 변수를 통해 범위 설정이 필요
  divs = soup.find('div', {'class':'newest-list'})

  titles = divs.findAll('a', {'class':'title'})
  artists = divs.findAll('a', {'class':'artist'})

  for i in range(50):
    if titles[i].text.strip().split('\n')[0] == '19금':
      # 뒤의 공백을 지우기
      title = titles[i].text.strip().split('\n')[-1].replace(' ', '')
    else:
      title = titles[i].text.strip()
    artist = artists[i].text.strip()
    genie.append((title,artist))


```

```python
print(genie)
print(len(genie))
```
