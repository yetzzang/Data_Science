# 3. 데이터 전처리

```python
park.dtypes
```

```python
park['공원면적'].head()
```

```python
# 공원면적비율 파생변수 만들기
# sqrt(공원면적) * 0.01
park['공원면적비율'] = park['공원면적'].apply(lambda x:np.sqrt(x) * 0.01)
park.head()
```

```python
park['소재지도로명주소'].isnull().sum()
```

```python
#도로명 주소는 입력되지 않고 지번 주소만 입력된 데이터
park.loc[(park['소재지도로명주소'].isnull()) & (park['소재지지번주소'].notnull())]
```

```python
# 도로명 주소와 지번 주소가 모두 입력된 데이터
park.loc[(park['소재지도로명주소'].notnull()) & (park['소재지지번주소'].notnull())]
```

```python
# 도로명 주소와 지번 주소가 모두 입력되지 않은 데이터
park.loc[(park['소재지도로명주소'].isnull()) & (park['소재지지번주소'].isnull())]
```

```python
# 도로명 주소가 NaN인 데이터를 지번 주소로 채움
park['소재지도로명주소'].fillna(park['소재지지번주소'], inplace=True)
```

```python
park['소재지도로명주소'].isnull().sum()
```

```python
park.head()
```

```python
# split(expand=True): 데이터프레임으로 분리되고 인덱싱과 슬라이싱이 가능
park['소재지도로명주소'].str.split(' ', expand=True)
```

```python
# 도로명 주소에서 '시도'만 추출해서 '시도' 파생변수를 만들고 park 데이터프레임에 추가
park['시도'] = park['소재지도로명주소'].str.split(' ', expand=True)[0]
park['읍면동'] = park['소재지도로명주소'].str.split(' ', expand=True)[2]
park.head()
```

```python
# 위도와 경도가 잘 못 입력된 데이터 추출
park_loc_error = park.loc[(park['위도'] <= 32) | (park['경도'] >= 132) | (park['시도'].isnull())]
```

```python
park_loc_error
```

```python
# 올바른 데이터
park_loc_notnull = park.loc[(park['위도'] > 32) & (park['경도'] < 132) & (park['시도'].notnull())]
```

```python
park_loc_notnull['시도'].value_counts()
```

```python
# '시도'에 '강원'이라고 잘못 표기된 데이터를 수정
park_loc_notnull['시도'][park_loc_notnull['시도'] == '강원'] = '강원도'
```

```python
park_loc_notnull['시도'].value_counts()
```

# 4. 시도별 공원 데이터 시각화

```python
ggplot(park_loc_notnull, aes(x='경도', y='위도', color='시도')) \
  + geom_point(size=0.5) \
  + theme(text=element_text(family='NanumGothic'), figure_size=(7,10))
```

```python
plt.figure(figsize=(8,10))
sns.scatterplot(data=park_loc_notnull, x='경도', y='위도', hue='시도')
```

```python
park_loc_notnull['공원구분']
```

```python
# 어린이 공원을 제외한 공원 분포
ggplot(park_loc_notnull[park_loc_notnull['공원구분'] != '어린이공원'], aes(x='경도', y='위도', color='공원구분', size='공원면적비율')) \
  + geom_point() \
  + theme(text=element_text(family='NanumGothic'), figure_size=(7,10))
```

```python
# 시도별 공원 비율
park['시도'][park['시도'] == '강원'] = '강원도'


park_do = pd.DataFrame(park['시도'].value_counts())
park_do
```

```python
# value_counts(normalize=True) : 전체 합계에 대한 비율이 계산
park_do_normalize = pd.DataFrame(park['시도'].value_counts(normalize=True))
park_do_normalize
```

```python
# value_counts(ascending=True) : 합계를 계산, 오름차순 정렬
park_do_ascending = pd.DataFrame(park['시도'].value_counts(ascending=True))
park_do_ascending
```

```python
# 시도별 합계 데이터와 비율 데이터를 병합
park_sido = park_do.merge(park_do_normalize, left_index=True, right_index=True).reset_index()
park_sido
```

```python
park_sido.columns = ['시도', '합계', '평균']
park_sido
```

```python
# stat = 'identity' : y축의 높이를 데이터의 값으로 bar 그래프 형태로 지정
ggplot(park_sido, aes(x='시도', y='합계')) \
    + geom_bar(stat='identity', fill='green') \
    + coord_flip() \
    + theme(text=element_text(family='NanumGothic'), figure_size=(10, 7))
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(data=park_sido, x='시도', y='합계')
```

```python
plt.figure(figsize=(12, 8))
plt.xticks(rotation=-45)
sns.barplot(data=park_sido, x='시도', y='합계')
```

```python
plt.figure(figsize=(12,8))
plt.yticks(rotation=45)
sns.barplot(data=park_sido, x='합계', y='시도')
```
