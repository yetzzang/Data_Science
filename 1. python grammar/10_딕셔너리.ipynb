#1. 딕셔너리 Dictionary 

* 대응관계를 나타내는 자료형으로 key와 value라는 것을 한 쌍으로 갖는 형태 
* 하나의 딕셔너리의 키값은 중복될 수 없음
* 하나의 딕셔너리의 값은 중복될 수 있음 
* 각각의 사물함의 키는 달라야함 BUT 그 안의 내용물은 달라야함 

###1-1 딕셔너리 만들기 

```python
#{키:값}

dic1 = {}
print(dic1)

dic2 = {'no':1, 'userid':'apple', 'name':'김사과', 'hp':'010-1111-1111'} 
print(dic2)
```

###1-2. 키로 값 찾기 

```python
dic2 = {'no':1, 'userid':'apple', 'name':'김사과', 'hp':'010-1111-1111'} 
print(dic2['userid'])
print(dic2['name'])
```

###1-3. 데이터 추가 및 삭제 

```python
dic3 = {1:'apple'}
print(dic3)

dic3[100] = 'orange' #키가 100이고 값이 'orange'인 값이 추가된다 b/c 키값중 100이 없으니까 
print(dic3)

dic3[50] = 'banana'
print(dic3)  #데이터를 추가한 순으로 추가됨!! 숫자나 알파벳 순이 아님 

dic3[1] = 'melon'  #같은 키값에 새로운 값을 replace 
print(dic3)

del dic3[50]
print(dic3)

print(type(dic3)) #<class 'dict>
```

###1-4. 딕셔너리 함수

```python
dic2 = {'no':1, 'userid':'apple', 'name':'김사과', 'hp':'010-1111-1111'} 

# keys() : only key만 리스트로 반환
print(dic2.keys())

# values() : only value 값만 리스트로 반환
print(dic2.values())

# items() : key와 value를 한쌍으로 묶어 튜플으로 반환
print(dic2.items())

# get() : key를 이용해서 value를 반환
print(dic2['userid']) 
#print(dic2.['age']) #KeyError: 'age'
print(dic2.get('userid')) 
print(dic2.get('age', '알수없음'))
#print(dic2.get('age', 'userid', 'hp')) #TypeError: get expected at most 2 arguments, got 3

# in : key가 딕셔너리 안에 있는지 확인
print('name' in dic2) #True
print('age' in dic2)  #False
```

###1-5 딕셔너리와 for 

```python
dic2 = {'no':1, 'userid':'apple', 'name':'김사과', 'hp':'010-1111-1111'}

for i in dic2:
  print(i, end=' ') #키값만 복사 

print()

for i in dic2.keys():
  print(i, end=' ')

print()

for i in dic2.values():
  print(i, end=' ') #value값만 복사 

print()

for i in dic2:
  print(dic2[i], end=' ')

print()

for i in dic2.keys():
  print(dic2.get(i), end=' ')
```

```python
#value값이 있는지 확인하고 싶을때 
for i in dic2.keys():
  if dic2.get(i) == 'apple':
    print('찾았다!')
```

```python
#딕셔너리 오름차순 / 알파벳 순서로 

print(sorted(dic2.items()))
```
