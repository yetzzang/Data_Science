#1. 반복문

###대입연산자 
* =, +=, -=, *=, /=, %= 
* 왼쪽에 저장된 값과 오른쪽의 값을 연산한 뒤에 다시 왼쪽에 대입 

````python
num = 10

num += 1
print(num)

num -= 1
print(num)

num *= 10
print(100)

num /= 10
print(num)

num %= 2
print(num)
````

###while문
```
while 조건식: 
  조건식이 True인 동안 반복할 문장 
```

````python
i = 1

while i <= 5:
  print('Hello Python!')
  i += 1

print('현재 i의 값 : {}'.format(i))
````

````python
# 1 ~ 10까지 합

i = 1
sum = 0

while i <= 10:
  sum += i
  i += 1

print('1부터 10까지의 합 : {}'.format(sum))
````

###for문
```
for 변수 in range(범위):
  반복할 문장 
```

````python
num = range(0,10,2) #(시작 인덱스, 종료 직전 인덱스, 증가값)0부터 시작하는 1씩 커지는 10개의 숫자 
print(num)
````

````python
for i in range(0, 10, 1):
  print(i)
````

````python
for i in range(10):  # (종료인덱스), 시작인덱스:0, 증가값: 1
  print(i, end=' ')
````

###문제
1~100까지 짝수의 총합을 구하는 프로그램을 작성해보자 

````python
sum = 0

for i in range(2, 101, 2):
  sum += i
print('1부터 100까지의 총합은 {}입니다'.format(sum))
````

````python
sum = 0

for i in range (1, 101):
  if i%2 == 0:
    sum += i
print('1부터 100까지의 총합은 {}입니다'.format(sum))
````

###문제
score 리스트에 저장된 점수가 60점 이사인 학생이 몇명인지 알아보는 프로그램을 작성해보자 
* score = [90, 30, 50, 60, 80, 70, 100, 40, 20, 10]

````python
score = [90, 30, 50, 60, 80, 70, 100, 40, 20, 10]
count = 0

for i in score:
  if i >= 60:
    count += 1

print('60점 이상인 학생수는 {}명 입니다'.format(count))
````

````python
count = 5

for i in range(1,6):
  for j in range(1,count+1):
    print('❤', end=' ')
  print()
  count -= 1
````

````python
# 선생님

for i in range(5):  #시작인덱스가 0임!!!!! 1로 헷갈리지 않기!!!!!!
  for j in range(i, 5):
    print('❤', end=' ')
  print()
````

###문제
2단부터 9단까지 구구단을 출력하는 프로그램을 작성해보자 (while문)

````python
num = 2
multiple = 1
result = 0

while num <= 9:
  print('{}단'.format(num))

  while multiple <= 10:
    result = num * multiple
    print('{} x {} = {}'.format(num, multiple, result))
    multiple += 1
  print()
  num += 1
  multiple = 1
  

 
````
