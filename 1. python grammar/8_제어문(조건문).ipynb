###2-2. if ~ else문
```
if 조건식 (결과가 True 또는 False가 나오는 문장) : 
  조건식이 True인 경우 실행할 문장
else:
  조건식이 False인 경우 실행할 문장
```

````python
age = int(input('나이를 입력하세요 : '))

if age > 19:  
  print('성인입니다')
else:
  print('미성년입니다')
print('입력한 나이 : ', age)
````

````python
age = int(input('나이를 입력하세요 : '))

if age > 19:  
  print('성인입니다')
  print('넷플릭스 성인 영화를 감상할 수 있습니다')
else:
  print('미성년입니다')
  print('디즈니를 보세요')
print('입력한 나이 : ', age)
````

###문제
* 국어, 영어, 수학 점수를 입력받아 총점, 평균, 학점을 구하는 프로그램 작성
  (단, 학점은 평균 기준이며 90 이상 A, 80이상 B, 70이상 C,  60이상 D, 나머지는 F)

````python
kor = int(input('국어 점수 : '))
math = int(input('수학 점수 : '))
eng = int(input('영어 점수 : '))

total = kor + math + eng
avg = float(total / 3)

print('총점은 {0:d}점이며 평균은 {1:5.2f}점입니다'.format(total, avg))
if avg >= 90:
  print('A학점입니다')
elif avg >= 80:
  print('B학점입니다')
elif avg >= 70:
  print('C학점입니다')
elif avg >= 60:
  print('D학점입니다')
else:
  print('F학점입니다')
````

````python
# 다른 버전
kor, eng, math = input('국어, 영어, 수학 점수를 입력하세요 : ').split(' ')
tot = int(kor) + int(eng) + int(math)
avg = tot / 3

#print('총점은', tot,'점 입니다')  #,(콤마)로 인해서 한칸씩 띄어쓰기가 됨 
#print('평균은 {0:5.2f}'.format(avg), '점 입니다')

print('총점은 %d점 입니다' %tot)
print('평균은 {0:5.2f}점 입니다'.format(avg))

grade = ''

if avg >= 90:
  grade = 'A'
elif avg >= 80:
  grade = 'B'
elif avg >= 70:
  grade = 'C'
elif avg >= 60:
  grade = 'D'
else:
  grade = 'F'

#print('학점은', grade, '학점입니다')
print('%s학점입니다' %grade)
````

### 논리 연산자 (AND)
* 양쪽 두 식의 값이 모두 True인 경우 결과가 True가 되는 연산자 


````python
# 로그인 프로그램 작성 
# 아이디 : admin, 비밀번호 : 1234

userid = input('아이디를 입력하세요 : ')
userpw = input('비밀번호를 입력하세요 : ')

if userid == 'admin' and userpw == '1234':
  print('로그인 되었습니다')
else:
  print('아이디 또는 비밀번호를 확인하세요')
````

###논리 연산자 (OR)
* 양쪽 두 식의 값 중 하나라도 True인 경우 결과가 True가 되는 연산자 

````python
# 주민등록번호에 의한 성별 판별
# 001011 - 3068518

juminNo = input('주민등록번호 13자리를 -없이 입력하세요') #0010113068518 #0으로 시작하기 때문에 String으로 받자 

if juminNo[6] == '1' or juminNo[6] == '3':
  print('남자')
elif juminNo[6] == '2' or juminNo[6] == '4':
  print('여자')
else:
  print('잘 못 입력하셨습니다')
````
