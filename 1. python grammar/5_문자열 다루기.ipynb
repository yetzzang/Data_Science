#1. 문자열 인덱싱 

* 문자열에 번호를 매겨 특정 문자를 찾을 수 있는 기능 

```python
# 각 alphabet 및 띄어쓰기에 인덱스 번호를 부여 
str = '문자열에 번호를 매겨 특정 문자를 찾을 수 있는 기능'  
print(str)
```

```python
print(str[0])
print(str[1])
print(str[12])
print(str[-1]) #마지막 문자 
print(str[-2]) 
```

#2. 문자열 슬라이싱 

* 문자열에서 원하는 부분을 추출하는 기능 (한글자 또는 여러글자)

```python
str = '문자열에서 원하는 부분을 추출하는 기능'
print(str) 
```

```python
print(str[0:5])
print(str[7:13])
print(str[6:-1])
print(str[-7:-1])
print(str[:13]) #앞에 아무것도 없다면 처음부터 
print(str[10:]) #뒤에 아무것도 없다면 끝까지
```

#3. 문자열 함수

* 문자열과 관련된 여러가지 기능(함수)

```python
str = '문자열 함수는 문자열과 관련된 여러가지 기능'
```

```python
# len() : 매개변수로 전달된 문자열의 갯수를 반환, 파이썬 표준 함수 (.을 찍지 않고 그냥 사용할 수 있는 함수)
print(len(str))
```

```python
# count() : 문자열에서 매개변수로 전달된 문자열의 출현 횟수를 반환
print(str.count('문'))
print(str.count('문자'))
print(str.count('문장'))
```

```python
# find() : 매개변수로 전달된 문자열이 처음 나오는 위치를 인덱스로 반환 
print(str.find('문'))
print(str.find('웩')) #없으면 -1이 나온다 
```

```python
# index() : 매개변수로 전달된 문자열이 처음 나오는 위치를 인덱스로 반환; find와 같지만 결과가 조금 다르게 나온다 
print(str.index('문'))
#print(str.index('웩')) #문자가 없으면 Error. ValueError: substring not found; 에러가 나면 나중에 에러처리를 통해 다른것을 해줄 수 있다. 
```

```python
# replace(찾을 문자열, 바꿀 문자열) : 문자열을 치환
print(str.replace('기능','함수')) 
```

```python
# upper() : 문자열을 대문자로 모두 변환 
# lower() : 문자열을 소문자로 모두 변환 
str = ' 문자열 function은 문자열과 관련된 여러가지 FUNCTION'
print(str.upper())
print(str.lower())

#한글자만 바꾸고 싶다면 인덱싱을 통해서 스트링 하나만 뺀 다음 클래스 바꿔주기 
```

```python
# lstrip() : 문자열의 왼쪽 공백을 제거 
str = '         Hello         Python          '
print(str)
print(str.lstrip())

# rstrip() : 문자열의 오른쪽 공백을 제거 
print(str.rstrip())

# strip() : 문자열의 양쪽 공백을 제거 
print(str.strip())

# 아예 공백을 다 없애주는 함수는??
# str.replace(' ','')
```

```python
# join() : 문자열에 특정 문자열을 삽입 
str = 'python'
print('😜'.join(str))  #나중에 스페이스를 문자열 사이사이에 삽입 시켜 주고 싶을때 
```
