#1. 파일 열기 

```
파일변수 = open('파일 이름', '파일 열기 모드')
```
* 파일 열기 모드: w(출력), a(추가 출력), r(입력)
* w는 지정된 경로에 파일이 없으면 파일을 만든 후 저장하고, 파일이 있으면 기존의 모든 데이터를 지우고 다시 저장
* a는 지정된 경로에 파일이 없으면 만든 후 저장하고, 파일이 있으면 기존의 데이터 뒤에 저장 
* r은 지정된 경로에 파일이 없으면 에러
* 출력시(w,a) 작업이 완료되면 반드시 close() 메소드를 사용해서 파일을 닫앙 정상적으로 데이터가 저장되고, 입력시(r) 파일을 닫지 않아도 정상적으로 처리 됨 
* 파일의 종류 구분: t(text), b(binary)
```
파일변수 = open('파일이름', 'r') # 텍스트파일 열기 모드
파일변수 = open('파일이름', 'rb') # 바이너리파일 열기모드  
```

````python
file = open('data.txt', 'wt')
for i in range(10):
  file.write('파일 열기 테스트 : ' + str(i) + '\n')  # 파일에 데이터를 써주는 역할
file.close()
print('data.txt 파일에 쓰기 완료!')
````

* 절대 경로 : 디스크 드라이브의 최상위(root) 디렉토리부터 텍스트 파일이 저장될 디렉토리까지의 경로
  * 예) D:WRyuzy\AI5\pythonWdata.txt
* 상대 경로: 현재 화면에 표시되는 소스 파일이 위치한 디렉토리부터 텍스트 파일이 저장될 디렉토리까지의 경로
  * 같은 디렉토리: 파일명(주로 윈도우), ./파일명(주로 리눅스, 맥), 예) data.txt, ./data.txt
  * 상위 디렉토리: ../파일명 예) ../data.txt
  * 하위 디렉토리: 디렉토리명/파일명, ./디렉토리명/파일명 예) dir/data.txt, ./dir/data.txt
  


````python
file = open('./dir/data.txt', 'wt')
for i in range(10):
  file.write('파일 열기 테스트: ' + str(i) + '\n')
file.close()
print('data.txt 파일에 쓰기 완료!')
````

#2. 파일 읽기

````python
# read() : 파일로부터 데이터를 읽어옴. 텍스트 모드와 바이너리 모드에서 가각 다른 방식으로 동작
# file.read(size) 또는 file.read()
# 텍스트의 경우 문자수, 바이너리의 경우 바이트 수

file = open('./dir/data.txt', 'rt')
data = file.read()
print('data.txt 파일 읽기 완료!')
print(data)
````

#3.with문 

* 자원을 획득하고 사용 후 반납해야 하는 경우 사용되는 문장 

````python
with open('./dir/output.txt', 'w') as file: #as는 별명 설정
 while True:
  data = input('단어를 입력하세요: ')
  if data.lower() == 'quit':
    break
  file.write(data + '\n')
print('output.txt 파일 저장완료') 
````

````python
class Dog:
  def __init__(self, name):  #생성자
    print(name, '객체 생성!')
  def __del__(self):
    print('객체 소멸!')
````

````python
Rucy = Dog('루시')
print(Rucy)
````

* 사이드 노트: del Rucy를 하게 되면, Rucy와 Dog의 커넥션을 끊어버림. 그렇게 되면 Dog는 파이썬 메모리 안에 미아가 되어버림. 나중에 파이썬의 쓰레기 콜렉터가 돌아다니다가 미아를 발견하면 자동으로 지움.

````python
del Rucy
````

````python
# readline(): 텍스트 파일을 한 줄씩 읽어서 처리, 파일이 종료되어 더이상 읽을수 없으면 빈 문자열('')을 일거온다 
file = open('./dir/data.txt', 'r')
lines = []

while True:
  line = file.readline()
  if not line:
    break
  if len(line.strip()) != 0:  #strip() 공백이 들어갔을 경우 공백 제거 
    print(line, end='')
    lines.append(line.strip())
file.close()
print()
print('data.txt 파일 읽기 완료!')
print(lines)
````

````python
# readlines() : 전체라인을 모두 읽어서 각 라인 단위로 리스트에 저장 
file = open('./dir/data.txt', 'r')
lines = file.readlines()
print(type(lines))
print(lines)
````
