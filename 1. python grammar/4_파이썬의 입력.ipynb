```python
name = input('이름을 입력하세요 : ')
print('%s님 안녕하세요!' %name)
```

```python
# split() : 인수로 전달된 문자를 기준으로 데이터를 나눠서 저장 
name, age, weight = input ('이름과 나이, 몸무게를 입력하세요 : ').split(' ')  #띄어쓰기를 기준으로 입력값이 각각의 variable에 저장됨 
print('{}님의 나이는 {}살이며, 몸무게는 {}kg입니다.'.format(name, age, weight))
```

```python
name, age, weight = input ('이름과 나이, 몸무게를 입력하세요 : ').split('/')  #나누는 값이 슬러시(/)를 기준으로 함 
print('{}님의 나이는 {}살이며, 몸무게는 {}kg입니다.'.format(name, age, weight))
```

```python
kor, eng, math = input('국어, 영어, 수학 점수를 입력하세요 : ').split(' ')
#print('총점 : ', kor + eng + math)

# int() : 문자열을 숫자(정수)로 변환
print('총점 : ', int(kor)+int(eng)+int(math))
#print('총점 : ', int(kor + eng + math)) # 우리가 원하는 결과 X, b/c 이 안에 있는 kor + eng + math가 먼저 나오기 때문에 

tot = int(kor)+int(eng)+int(math)
avg = float(tot / 3)

#print('총점 : ', tot)
#print('평균 : ', avg) # 나누기를 쓰면 정수로 나누기를 하더라도 소수점을 나옴 

print('총점 : {0}  평균 : {1:5.2f}'.format(tot, avg))
```
