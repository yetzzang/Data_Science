###문제
* 첫번째 두번째 숫자를 입력 받아 그 사이 숫자들의 총합을 구한다  

````python
def func3(start, end):
  sum = 0
  for i in range(start, end+1, 1):
    sum += i
  print("%d부터 %d까지의 합 : %d" % (start, end, sum))
````

````python
func3(1, 10)
func3(1, 100)
````

###1-5 매개변수의 개수가 가변적인 함수
```
def 함수명(*변수):
  함수가 호출되면 실행할 문장
  ...
  return 값
```

```
변수 = 함수명(값1) # 호출
변수 = 함수명(값1, 값2 ...) # 호출
```
> 매개변수는 함수안에서 튜플형태로 사용

````python
def func7(*args):
  #print(args)
  result = 0
  for i in args:
    result += i
  return result
````

````python
print(func7(10))
print(func7(10, 30, 20, 70, 50))
````

### 1-6 매개변수를 딕셔너리 구조로 전달받는 함수 
```
def 함수명(**변수):
  함수가 호출되면 실행할 문장
  ...
  return 값
```

```
변수 = 함수명(변수1=값1, 변수2=값2, ...) # 호출
```


````python
def func8(**args):
  return args 
````

````python
user = func8(userid='apple', name='김사과', age=20)  #이 값들이 딕셔너리 형태로 저장됨 
print(user)
print(type(user))
````

###1-7. 리턴값이 여러개인 함수 
```
def 함수명(**변수):
  함수가 호출되면 실행할 문장
  ...
  return 값1, 값2, ...
```

```
변수1, 변수2, ... = 함수명(값1, 값2, ...)  #호출 
```

````python
def func9(num1, num2):
  return num1+num2, num1-num2, num1*num2, num1/num2
````

````python
result1, result2, result3, result4 = func9(10, 3)
print(result1, result2, result3, result4)
print(func9(10, 3)) #다른가??? 
````

````python
result1 = func9(10, 3)
print(result1)

_, _, result1, _ = func9(10, 3)  #리턴 값에 맞게 변수를 저장해야 하는데 필요가 없을때 언더바로 없애줌 
print(result1)
````
