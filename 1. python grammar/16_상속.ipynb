#1. 상속

* 기존의 정의해둔 클래스의 기능을 그래도 물려받음
* 기존 클래스에 기능 일부를 추가하거나, 변경하여 새로운 클래스를 정의함 
* 코드를 재사용할 수 있게 됨
* 상속을 받고자 하는 대상인 기존 클래스를 부모, parent, super, base 
* 상속 받는 새로운 클래스를 자식, child, sub라고 칭함 

```python
class Animal:
  def __init__(self, name, age):
    self.name = name
    self.age = age
  
  def eat(self, food):
    print('{}는 {}를 먹습니다'.format(self.name, food))
  
  def sleep(self, hour):
    print('{}는 {}시간 동안 잡니다'.format(self.name, hour))

  
```

```python
animal = Animal('동물', 5)
animal.eat('먹이')
animal.sleep(5)
```

```python
# 상속 하는 법

class Dog(Animal): #Animal 클래스를 상속
  pass
```

```python
#Rucy = Dog()  #TypeError: __init__() missing 2 required positional arguments: 'name' and 'age'  #이미 Dog는 Animal을 상속 받은 상태 > parameter 필요
Rucy = Dog('루시', 12)
Rucy.eat('사료')
Rucy.sleep(8)
```

#2. 메소드 오버라이딩 

* 부모 클래스의 method를 재정의
* 자식 클래스의 인스턴스로 호출시, 재정의된 메소드가 호출됨 

```python
class Animal:
  def __init__(self, name, age):
    self.name = name
    self.age = age
  
  def eat(self, food):
    print('{}는 {}를 먹습니다'.format(self.name, food))
  
  def sleep(self, hour):
    print('{}는 {}시간 동안 잡니다'.format(self.name, hour))
```

```python
#메소드 오버라이딩
class Dog(Animal):
  def eat(self, food, cnt):
    print('{}는 {}를 {}개 아주 맛있게 먹습니다!'.format(self.name, food, cnt))
```

```python
Rucy = Dog('루시', 12)
#Rucy.eat('사료') #TypeError 
Rucy.eat('사료', 3)
Rucy.sleep(8)
```

#3. 다중 상속

```python
class Animal:
  def __init__(self, name, age):
    self.name = name
    self.age = age
  
  def eat(self, food):
    print('{}는 {}를 먹습니다'.format(self.name, food))
  
  def sleep(self, hour):
    print('{}는 {}시간 동안 잡니다'.format(self.name, hour))
```

```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def study(self, hour):
    print('{}는 {}시간 동안 공부합니다'.format(self.name, hour))
```

```python
class Kim(Animal, Human):
  pass
```

```python
kim = Kim('김사과', 20)  #다중상속시 부모의 __init__의 parameter들이 무조건 같아야함 
kim.eat('밥')
kim.study(4)
```
