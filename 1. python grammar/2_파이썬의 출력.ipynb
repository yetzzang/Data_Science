#2. 출력 서식 지장하기

1. %d(정수;integer), %f(실수;real number), %s(문자열;character): print 안에서 이걸 쓰면 정수/실수/문자열로 출력
2. 서식을 제외한 문자는 입력한 그대로 출력이 된다
3. 서식을 지정해 출력할 데이터가 2개 이상인 경우 (소괄호)로 데이터를 묶어야 한다  

```python
print('숫자 : 100')
print('숫자 : %d' %100) #%d자리가 숫자 100으로 바뀜 
print('숫자 : %d' %100.5) #정수가 들어가야 하는 자리이기 때문에 100으로 자동으로 바뀜

print('실수 : %f' %100.5)

print('문자열 : %s' %'Python')
print ('문자열 : %s' %100)

#5 + 3 = 8
#단, 출력 서식을 이용하고 결과는 실제 덧셈 결과를 출력 
print('%d + %d = %d' %(5,3,5+3))
#print('%d' %5 '+ %d' %3 '= %d' %5+3)  #안되는 이유는 formatter와 parameter가 구분지어져야 함 ex) 왼쪽에 formatter 오른쪽에 parameter; 이경우엔 formatter와 parameter가 섞여있기 때문에 에러가 뜸  
#print ('%d' %5 ,'+ %d' %3 ,'= %d' %5+3)  #위와 같은 이유로 에러가 뜸 
```

```python
print('%5d' %100) #총 다섯자리를 확보; 그래서 앞에 두자리를 띄우고 100을 출력 
print('%5d' %10000)
print('%5d' %100000) #그냥 밀어냄
print('%05d' %100) #나머지 빈칸을 0으로 채워줌 
```

```python
print('%6.2f' %123.45) #점까지 포함해서 6자리 수이고 2자리를 소수점으로
print('%6.2f' %123.456)
print('%6.2f' %123.4)
```

```python
print('%5s' %'abc')
```

```python
# 5 / 3 = ?
print('%d / %d = %d' %(5,3,5/3)) #결과를 %d로 출력하면 정수로 출력
print('%d / %d = %f' %(5,3,5/3)) #결과를 %f로 출력하면 실수로 출력 
```

#3. format 함수와 출력 인덱스 사용  

* 인덱스 : 숫자 순서 매기는 것; 01234 (0부터 시작함) 

```python
print('{0} {1}'.format('김사과','반하나')) #앞에 있는 문자열에 .format이라는 함수를 사용하여 순서를 매김 
print('{1} {0}'.format('김사과','반하나'))
print('{1} {2} {0}'.format('김사과',' ','반하나'))
```

```python
print('{0:3s}님은 {1:03d}살 입니다'.format('김사과',20))   #0번째 인덱스에는 세자리의 string이 들어가고 첫번째 인덱스에는 세자리 정수가 들어감. 그러나 빈자리는 0으로 채움 
print('{0}님의 키는 {1:6.2f}cm입니다'.format('반하나',160.5)) #0번째 인덱스에는 첫번째가 들어가고 1번째 인덱스에는 6.2f의 실수가 들어감 (6자리 숫자의 두자리 소수점). 그래서 160.5 앞에 한자리가 띄어주고 뒤에는 0으로 채워짐 
```
