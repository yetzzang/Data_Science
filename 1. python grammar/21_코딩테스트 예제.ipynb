#1. 수 정렬하기

https://www.acmicpc.net/problem/2750

```python
n = int(input())
data = list()

for i in range(n):
  # num = int(input())
  # data.append(num)
  data.append(int(input()))

data.sort()

for i in range(len(data)):
  print(data[i])
```

#2. 음계

https://www.acmicpc.net/problem/2920

```python
n = input().split()
a_temp = n.sort()
de_temp = n.sort(reverse = True)

output = ''
ascending = ''
descending = ''

for i in range(8):
    output += n[i]
    ascending += a_temp[i]
    descending += de_temp[i]

if output == ascending:
  print('ascending')
elif output == descending:
  print('desending')
else:
  print('mixed')
```

```python
num = list(map(int, input().split(' ')))

ascending = True
desencding = True

for i in range(1,8):
  if num[i] > num[i-1]:
    descending = False
  elif num[i] < num[-1]:
    ascending = False

if ascending:
  print('ascending')
elif desceding:
  print('descending')
else:
  print('mixed')
```
