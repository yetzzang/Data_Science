# 1. 너비 우선 탐색 Breadth-First Search (BFS)

### 1-1. BFS
* 대표적인 그래프 탐색 알고리즘 
* vertex들과 같은 레벨에 있는 노드들을 먼저 탐색하는 방식 
  * 예시) 트리 그래프에서 같은 레벨에 있는 애들을 옆으로 하나씩 가면서 탐색 
* 한 단계씩 내려가면서 해당 노드와 같은 레벨에 있는 노드들을 먼저 순회함 

```python
graph = dict()

graph['A'] = ['B', 'C']
graph['B'] = ['A', 'D']
graph['C'] = ['A', 'G', 'H', 'I']
graph['D'] = ['B', 'E', 'F']
graph['E'] = ['D']
graph['F'] = ['D']
graph['G'] = ['C']
graph['H'] = ['C']
graph['I'] = ['C', 'J']
graph['J'] = ['I']

graph
```

# 2. BFS 알고리즘 구현 

* 큐 자료구조를 활용
* need_vist 큐, visited 큐 - 두개의 큐를 생성 

```python
# pop() 복습 
data = [1,2,3, 4]
data.pop()
print(data)
data.pop(0) # data[0]이 pop됨 
print(data)
```

```python
# extend() 복습 
data = [1,2,3,4]
data.extend([4,5])
print(data)
```

```python
def bfs(graph, start_node):
  visited, need_visit = list(), list()

  need_visit.append(start_node)

  while need_visit:
    node = need_visit.pop(0)
    if node not in visited:
      visited.apped(node)
      need_visit.extend(graph[node])
  return visited 
```
