# 1. 재귀 호출 Recursive Call

* 함수 안에서 동일한 함수를 호출하는 형태
* 여러 알고르짐, 고급 정렬 알고리즘 작성시 자주 사용됨 

### 1-1. 재귀 호출 분석 
* factorial ! - 3! = 3 * 2 * 1
* this also means n! = n * (n-1)!

### 1-2. 규칙 
* n! = n * (n-1)!
 1. 함수(n)은 n>1 이면 return n * 함수(n-1); return에서 함수를 끈임없이 recursive하게 호출함 
 2. 함수(n)은 n=1이면 return n 

### 1-3. 검증
* 2! 
  * 함수(2)이면 2>1이므로 2*함수(1)
  * 함수(1)은 1 이므로 return 2 * 1, 답은 2
* 3!
  * 함수(3)이면 3>1이므로 3 * 함수(2)
  * 함수(2)는 위에 의해 2!이므로 return 2*1 = 2
  * 그래서 3*함수(2) = 3 * 2* 1 = 6

```python
def factorial(num):
  if num > 1:
    # recursive call of function itself 
    return num * factorial(num - 1)
  else:
    return num 
```

```python
factorial(3), factorial(4), factorial(5)
```

### 1-4. recursive call 재귀 호출의 시간 복잡도

* factorial(n)은 n-1번의 factorial()을 호출해서 곱셈을 함
  * n-1번 반복문을 호출한 것과 동일
  * factorial()함수를 호출할때마다 지역변수 n이 생성됨; n번이 return 된다는 뜻
  * 시간 복잡도는 O(n-1)이므로 O(n) 

### 1-5. 재귀 호출의 전형적인 예
* 함수는 내부적으로 스택처럼 관리 
> 파이썬에서는 재귀함수의 깊이가 (한번에 호출되는) 횟수가 1000번 이상이 되면 에러가 난다. 

### 문제
회문 (Palindrome)은 "순서를 거꾸로 읽어도 제대로 읽은 것과 같은 단어와 문장을 의미"
회문을 판별할 수 있는 함수를 리스트 슬라이싱을 활용하여 만들어보자.
* ex) 'abba' 'kayak', 'reviver', 'madam'
* 재귀함수를 사용
* 회문이면 True, 아니면 False 반환 


```python
def palindrome(string):
  if len(string) <= 1:
    return True
  if string[0] == string[-1]:
    return palindrome(string[1:-1])
  else:
    return False
```

```python
palindrome('kayak'), palindrome("tomato")
```

### 문제
* 정수 n을 입력받아 아래와 같이 처리하는 프로그램을 만들어보자
* n이 홀수면 3 * n + 1
* n이 짝수면 n을 2로 나눔
* 이렇게 계속 진행해서 n이 결국 1이 될때까지 위를 반복 

```python
def calc(n):
  print(n)
  if n == 1:
    return n
  if n % 2 == 1:
    return(calc(3*n+1))
  else:
    return (calc(int(n/2))) # 2로 나누면 float가 되서 int()를 씌움 
```

```python
calc(21)
```
