```python
df = pd.read_csv("https://bit.ly/ds-korean-idol")
df2 = pd.read_csv("https://bit.ly/ds-korean-idol-2")
```

```python
df_right = df2.drop([1,3,5,7,9], axis=0)
```

```python
df_right
```

```python
df_right = df_right.reset_index(drop=True)
df_right
```

```python
# 합쳐는지지만 문제가 생김
# real-life에선 인덱스가 아름답지는 않다
pd.concat([df, df_right], axis=1)
```

```python
# merge() : 특정 고유한 키 (unique id)값을 기준으로 병합
# on : unique id를 선택
# how: 병합의 기준 (left, right, inner, outer)
# left일 경우, 왼쪽의 데이터는 모두 확인하고, 붙혀질 오른쪽 데이터가 빌 경우, 그 경우는 NaN을 채움
# right일 경우, 반대
pd.merge(df, df_right, on="이름", how="left")
```

```python
df_right = df_right.append({"이름":"김사과", "연봉":2800, "가족수":4}, ignore_index=True)
```

```python
# 오른쪽의 경우로 맞춰졌기 때문에,
# 김사과의 경우엔 df의 15번째가 NaN으로 채워지고
# df_right의 사라진 인덱스의 대한 경우는 df의 데이터가 사라짐
pd.merge(df, df_right, on="이름", how="right")
```

```python
# inner의 경우 : 교집합; 둘의 데이터프레임에 교집합 되는 경우만 합침
# outer: 합집합
pd.merge(df, df_right, on="이름", how='inner')
```

```python
pd.merge(df, df_right, on="이름", how="outer")
```

```python
# 데이터는 같지만 그 이름이 다를 경우
df_right.columns = ["성함", "연봉", "가족수"]
```

```python
# pd.merge(df, df_right, on="이름", how="outer")
```

```python
pd.merge(df, df_right, left_on="이름", right_on = "성함", how="outer")
```

# 25. 원 핫 인코딩 (One-Hot-Encoding)

* 원 핫 인코딩은 한개의 요소는 True, 나머지 요소는 False로 만들어주는 기법

```python
blood_map = {"A":0, "B":1, "AB":2, "O":3}
df["혈액형_code"] = df["혈액형"].map(blood_map)
df.head()
```

```python
df["혈액형_code"].value_counts()
```

* df["혈액형_code"]를 머신러닝 알고리즘에 넣어 데이터를 예측하려고 지시를 한다면 컴퓨터는 갑들간의 관계를 스스로 형성하게 됨
* 만약 B형은 1, AB형은 2라는 값을 가지고 있을때, 컴퓨터는 'B형 + AB형 = O형"라고 잘못 된 관계를 맺을 수 있게됨
* 별도의 column을 형성해주고 1rodml column에는 True, 나머지는 False라고 넣어줌으로 'A, B, AB, O형"의 관계는 독립적이다라는 카테고리로 표현해주는 방식

```python
pd.get_dummies(df["혈액형_code"])
```
