### HTTP

* HTTP란?
    * HyperText(HTML) Transfer Protocol
    * simple 'request-response' protocol layerd on TCP/IP
    * IMPORTANT:
        * HTTP로 요청을 보내면 __반드시__ 응답을 보낸다
        * HTML 문서로 주고 받는다
        * 이때, 어떤 브라우저나 서버가 사용되는지는 상관이 없다
>
* TCP/IP란?
    * 통신을 주고 받는 방법 중 하나
      >
* IP
    * 패킷으로 쪼갠다
        * what does it mean? 인터넷은 점유식이 힘들다. therefore we 쪼개다 패킷 so we can make them possible to use Internet
    * 점유식? 누군가와 소통 중이면 다른이와의 소통 접근은 제한된다
    * 패킷이란?
        * 헤더
        * 상대방에게 보내는 것
>
* TCP
    * 신뢰성 있는 연결은 보장
    * IP를 패킷을 보냈을 때 어떻게 알아?
    * HTTP에선 요청을 보냈을 때 응답을 안하는 서버는 없다
>
* HTML이란?
    * HyperText Markup Language
    * 가상의 링크가 존재 -> 다른 곳으로 이동
        * ex) 책에서 갑자기 chapter 4로 가고 싶다? 앵커로 슝 이동
    * __링크__: '앵커'라고도 한다
    * 매우 중요하다
    * note: 구글 정보 검색이 좋은 이유는 링크 때문이다
        * 링크 활용 - 이 서비스가 이 서비스로 링크했다
> IMPORTANT TO KNOW:
> HTTP: 요청은 반드시 응답한다
> HTML: 링크 주소 해석 능력이 중요하다   

```python
# TCP/IP => Bytes(00101010100)
# HTTP => text/html
#        Text => Bytes (요청)
#        Bytes => Text (응답)
```

* 알아야 하는 용어들:
    * __인코딩 encoding__ 이란?
        * 데이터를 특정 형식 또는 규칙에 따라 다른 형식으로 변환하는 과정
        * 변환된 데이터는 원래의 형태와는 다른 형식으로 표현
        * 데이터를 저장, 전송, 처리 하기 위해 사용
        * ex) 텍스트 데이터를 UTF-8/아스키코드로 변환
          >
    * __디코딩 decoding__ 이란?
        * 인코딩된 데이터를 원래의 형식 또는 규칙에 따라 다시 원래의 형태로 변환
        * 디코딩해서 다시 읽을 수 있는 형태로
        * ex) UTF-8/아스키 --> text
          >
    * __파싱(parsing)__ 이란?
        * 입력 데이터를 의미 있는 부분으로 분해하고 해석하는 과정
        * HTML 문서 형식을 읽고 이해
        * 웹 스크래핑이나 웹 크롤링과 같은 작업에서는 웹 페이지의 내용을 파싱하여 필요한 정보를 추출하거나 원하는 데이터를 가져오는 데 사용
          >
    * __언파싱 (unparese)__ 이란?
        * 구문 분석 (파싱)된 텍스트 또는 데이터를 원래의 텍스트 형식으로 되돌리는 것을 의미
          >
    * __escape__ 이란?
        * 특수 문자나 예약 문자를 HTML Entity로 변환하는 과정
            * HTML Entity: 특정 의미를 가지거나 특정한 동작을 수행할 수 있는 문자 
        * HTML 문서에서 해당 문자를 표시할 때에도 그 문자가 HTML 태그나 특수 기호로 인식되지 않고 원래의 의도대로 표시된다
        * ex) <는 HTML에서 태그로 시작하는 기호로 사용, but 텍스트로 표시하기 위해서는 이스케이프 처리 과정 필요
          >
    * __unescape__ 이란?
        * 이스케이프 된 HTML Entity를 원래의 문자로 다시 변환

```python
# https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=%EC%88%98%EC%A7%80
url = 'https://search.naver.com/search.naver'
params = {
    'where':'nexearch',
    'sm':'top_hty',
    'fbm':'0',
    'ie':'utf8',
    'query':'수지'
}

resp = request('GET', url, params=params)
resp.status_code, resp.reason
# ua를 바꿔야 내가 보는 웹페이지랑 진짜랑 같다 
```

```python
resp.headers, resp.request.headers, resp.request.url, resp.request.body
```

```python
resp.text
```

```python
re.findall(r'<a target="_blank" href="([^"]+?)" class="link_tit"[^>]+?>(.+?)</a>',resp.text)
```

> 주의 사항: element는 DOM이라고 해서 브라우저가 이미 해석을 끝 마친것; 우리가 봐야하는 것은 network 안에 있는 response이다. 따라서, 둘 사이에 차이가 있을 수 있다
