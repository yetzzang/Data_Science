### 3-1 impute
* 2개 이상의 column을 한번에 함수 또는 수식을 처리할 때 사용
* sklearn 안에 포함

```python
from sklearn.impute import SimpleImputer
```

```python
# mean, median, most_frequent
# constant: 특정값 예) SimpleImputer(strategy='constant', fill_value=값)
imputer = SimpleImputer(strategy='mean')
```

```python
imputer.fit(train[['Age', 'Pclass']]) # fit()을 통해 결측치에 대한 mean 학습을 진행
```

```python
result = imputer.transform(train[['Age', 'Pclass']])
# transform()은 결측치에 대한 값 변환을 적용
```

```python
result
```

```python
train[['Age', 'Pclass']].describe()
```

```python
train['Embarked'].fillna('C').describe()
```

```python
imputer = SimpleImputer(strategy='most_frequent')
result = imputer.fit_transform(train[['Embarked', 'Cabin']])
```

```python
train[['Embarked', 'Cabin']] = result
```

```python
train[['Embarked', 'Cabin']].isnull().sum()
```

```python
train[['Embarked', 'Cabin']].describe()
```

# 5. One Hot Encoding

```python
train = pd.read_csv('https://bit.ly/fc-ml-titanic')
train.head()
```

```python
train.Embarked.value_counts()
```

```python
train.Embarked.isnull().sum()
```

```python
# NaN 값을 'S'로 변환
train.Embarked = train.Embarked.fillna('S')
```

```python
train.Embarked.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
train['Embarked_num'] = LabelEncoder().fit_transform(train.Embarked)
```

```python
train['Embarked_num'].value_counts()
```

### 'S'=2, 'Q'=1, 'C'=0
* 머신러닝은 위 데이터의 관계성을 찾아 Q + Q = S 라고 학습을 할 수 있음
* 따라서 독립적인 데이터는 별도의 column으로 분리하고 각각의 컬럼에 해당 값에만 True, 나머지를 False 값을 갖게 하는 기술을 적용

```python
pd.get_dummies(train.Embarked_num)[:5]
```
